## 1. í™˜ê²½ ì„¤ì • ë° ë¼ì´ë¸ŒëŸ¬ë¦¬ ì„í¬íŠ¸

```python
import h5py
import numpy as np
import pandas as pd
from pathlib import Path
from tqdm import tqdm
import warnings

warnings.filterwarnings('ignore')
pd.set_option('display.width', 100)
```

## 2. ë°ì´í„° í†µê³„ ë¶„ì„

ëª¨ë“  `kspace` í´ë”ë¥¼ ìˆœíšŒí•˜ì—¬ ê° `.h5` íŒŒì¼ì˜ `kspace` í…ì„œ í˜•íƒœ ì •ë³´ë¥¼ ìˆ˜ì§‘í•˜ê³  ë¶„ì„í•©ë‹ˆë‹¤.

```python
base_dir = Path('data1')
kspace_dirs = [
    p for p in base_dir.rglob('*') 
    if p.is_dir() and p.name == 'kspace'
]

print(f"ë°œê²¬ëœ kspace í´ë”ë“¤:\n{[str(d) for d in kspace_dirs]}")
```

```python

all_files_info = []
for kspace_dir in kspace_dirs:
    h5_files = list(kspace_dir.glob('*.h5'))
    if not h5_files:
        continue

    # í´ë” ê²½ë¡œë¥¼ ë” ëª…í™•í•˜ê²Œ í‘œì‹œí•˜ë„ë¡ ìˆ˜ì •
    pretty_folder_name = "/".join(kspace_dir.parts[-3:])
    for h5_file in tqdm(h5_files, desc=f"Scanning {pretty_folder_name}"):
        try:
            with h5py.File(h5_file, 'r') as f:
                if 'kspace' not in f:
                    continue
                slices, coils, H, W = f['kspace'].shape
                all_files_info.append({
                    'folder': pretty_folder_name, # ê·¸ë£¹í™”ë¥¼ ìœ„í•œ í´ë”ëª…
                    'filename': h5_file.name,
                    'slices': slices,
                    'coils': coils,
                    'H': H,
                    'W': W
                })
        except Exception as e:
            print(f"Error processing {h5_file}: {e}")

df_stats = pd.DataFrame(all_files_info)
```

## 3. `acc4` vs `acc8` ìŠ¤ìº” ê²€ì¦

### ê²€ì¦ ë°©ë²•ë¡  ì œì•ˆ
ë‹¨ìˆœíˆ k-space ë°ì´í„°ë¥¼ ëª¨ë‘ ë¹„êµí•˜ëŠ” ê²ƒì€ ë¹„íš¨ìœ¨ì ì´ë¯€ë¡œ, ë‹¤ìŒê³¼ ê°™ì€ í•©ë¦¬ì ì¸ ë‘ ë‹¨ê³„ ê²€ì¦ ë°©ë²•ì„ ì‚¬ìš©í•©ë‹ˆë‹¤.

1.  **ë©”íƒ€ë°ì´í„°(Metadata) ë¹„êµ**: HDF5 íŒŒì¼ì˜ `attrs`ì— ì €ì¥ëœ ë©”íƒ€ë°ì´í„°ë¥¼ ë¹„êµí•©ë‹ˆë‹¤. ë™ì¼í•œ í™˜ìì˜ ìŠ¤ìº”ì´ë¼ë©´ `patient_id`ì™€ ê°™ì€ ì£¼ìš” ì†ì„±ë“¤ì´ ì¼ì¹˜í•´ì•¼ í•©ë‹ˆë‹¤. ì´ ë°©ë²•ì´ ê°€ì¥ ë¹ ë¥´ê³  íš¨ìœ¨ì ì…ë‹ˆë‹¤.

2.  **ë°ì´í„° ìƒ˜í”Œë§ ë¹„êµ**: ë©”íƒ€ë°ì´í„° í™•ì¸ í›„, ì¶”ê°€ì ì¸ ê²€ì¦ì„ ìœ„í•´ ì‹¤ì œ k-space ë°ì´í„°ë¥¼ ë¹„êµí•©ë‹ˆë‹¤. `acc4`ì™€ `acc8` ë§ˆìŠ¤í¬ê°€ ëª¨ë‘ 1ì¸, ì¦‰ **ê³µí†µìœ¼ë¡œ ìƒ˜í”Œë§ëœ ì˜ì—­ì˜ ë°ì´í„°ëŠ” ì™„ì „íˆ ì¼ì¹˜**í•´ì•¼ í•©ë‹ˆë‹¤. ëª¨ë“  ìŠ¬ë¼ì´ìŠ¤ë¥¼ ë¹„êµí•  í•„ìš” ì—†ì´, ì¤‘ì•™ë¶€ì˜ ëŒ€í‘œ ìŠ¬ë¼ì´ìŠ¤ í•˜ë‚˜ì—ì„œ ê³µí†µ ìƒ˜í”Œë§ëœ ë°ì´í„°ê°€ ì¼ì¹˜í•˜ëŠ”ì§€ í™•ì¸í•˜ì—¬ ë°ì´í„° ì§€ë¬¸(fingerprint)ì²˜ëŸ¼ ì‚¬ìš©í•©ë‹ˆë‹¤.

```python
import h5py
import numpy as np
from pathlib import Path

def verify_scans_data_only(file_acc4: Path, file_acc8: Path) -> (bool, str):
    """
    ë©”íƒ€ë°ì´í„° ì—†ì´ ì˜¤ì§ ë°ì´í„°ì˜ Shapeê³¼ ìƒ˜í”Œ ê°’ë§Œìœ¼ë¡œ
    ë‘ h5 íŒŒì¼ì´ ë™ì¼í•œ ì›ë³¸ ìŠ¤ìº”ì¸ì§€ ê²€ì¦í•©ë‹ˆë‹¤.
    """
    if not file_acc4.exists() or not file_acc8.exists():
        return False, "íŒŒì¼ ì—†ìŒ"

    try:
        with h5py.File(file_acc4, 'r') as f4, h5py.File(file_acc8, 'r') as f8:
            kspace4 = f4['kspace']
            kspace8 = f8['kspace']

            # 1ë‹¨ê³„: k-space í…ì„œì˜ Shape ì „ì²´ ë¹„êµ
            # (slices, coils, H, W) ì°¨ì›ì´ ëª¨ë‘ ë™ì¼í•´ì•¼ í•©ë‹ˆë‹¤.
            if kspace4.shape != kspace8.shape:
                return False, f"k-space Shape ë¶ˆì¼ì¹˜. F1:{kspace4.shape}, F2:{kspace8.shape}"

            # 2ë‹¨ê³„: ê³µí†µ ìƒ˜í”Œë§ ì˜ì—­ì˜ 'ì „ì²´' ë°ì´í„° ë¹„êµ
            mask4 = f4['mask'][:]
            mask8 = f8['mask'][:]
            
            # ë‘ ë§ˆìŠ¤í¬ì—ì„œ ê³µí†µìœ¼ë¡œ ìƒ˜í”Œë§ëœ k-space ì£¼íŒŒìˆ˜ ì¶•(W)ì˜ ì¸ë±ìŠ¤ë¥¼ ì°¾ìŠµë‹ˆë‹¤.
            common_cols = np.where((mask4 == 1) & (mask8 == 1))[0]

            if len(common_cols) == 0:
                return False, "ê³µí†µ ìƒ˜í”Œë§ ì˜ì—­ ì—†ìŒ"

            # 'ëª¨ë“  ìŠ¬ë¼ì´ìŠ¤'ì™€ 'ëª¨ë“  ì½”ì¼'ì— ëŒ€í•´ ê³µí†µ ì˜ì—­ ë°ì´í„° ì¶”ì¶œ
            # Ellipsis(...)ëŠ” ëª¨ë“  ì´ì „ ì°¨ì›(slice, coil, H)ì„ ì˜ë¯¸í•©ë‹ˆë‹¤.
            kspace4_common = kspace4[..., common_cols]
            kspace8_common = kspace8[..., common_cols]
            
            # ë³µì†Œìˆ˜ ë°°ì—´ ì „ì²´ë¥¼ ë¹„êµí•©ë‹ˆë‹¤.
            if not np.allclose(kspace4_common, kspace8_common):
                return False, "k-space ìƒ˜í”Œ ë°ì´í„° ë¶ˆì¼ì¹˜"

    except Exception as e:
        return False, f"íŒŒì¼ ì²˜ë¦¬ ì¤‘ ì˜¤ë¥˜: {e}"
    
    return True, "ê²€ì¦ ì„±ê³µ (Shape ë° k-space ìƒ˜í”Œ ë°ì´í„° ì¼ì¹˜)"
```

```python
print("--- Train ë°ì´í„°ì…‹ acc4 vs acc8 ê²€ì¦ ì‹œì‘ ---\n")
train_kspace_dir = base_dir / 'train/kspace'
verification_results = []
all_verified = True

scan_types = {
    'brain': range(1, 86),
    'knee': range(1, 86)
}

for organ, indices in scan_types.items():
    print(f"[{organ.capitalize()} Scans ê²€ì¦]")
    num_failed = 0
    for i in tqdm(indices, desc=f"Verifying {organ}"):
        file4 = train_kspace_dir / f"{organ}_acc4_{i}.h5"
        file8 = train_kspace_dir / f"{organ}_acc8_{i}.h5"
        
        # ìˆ˜ì •ëœ í•¨ìˆ˜ í˜¸ì¶œ
        is_same, reason = verify_scans_data_only(file4, file8)
        result = {
            'organ': organ,
            'index': i,
            'verified': is_same,
            'reason': reason
        }
        verification_results.append(result)
        if not is_same:
            all_verified = False
            num_failed += 1
            print(f"  - ê²€ì¦ ì‹¤íŒ¨: {organ}_acc*_{i}.h5, ì´ìœ : {reason}")
    
    if num_failed == 0:
        print(f"  >> ëª¨ë“  {organ} ìŠ¤ìº” í˜ì–´ê°€ ì„±ê³µì ìœ¼ë¡œ ê²€ì¦ë˜ì—ˆìŠµë‹ˆë‹¤.\n")
    else:
        print(f"  >> {organ} ìŠ¤ìº” ì¤‘ {num_failed}ê°œ í˜ì–´ê°€ ê²€ì¦ì— ì‹¤íŒ¨í–ˆìŠµë‹ˆë‹¤.\n")

print("--- ê²€ì¦ ì™„ë£Œ ---")
if all_verified:
    print("âœ… ëª¨ë“  ìŠ¤ìº” í˜ì–´ê°€ ë™ì¼í•œ ì›ë³¸ ë°ì´í„°ì„ì„ ì„±ê³µì ìœ¼ë¡œ í™•ì¸í–ˆìŠµë‹ˆë‹¤.")
else:
    print("âŒ ì¼ë¶€ ìŠ¤ìº” í˜ì–´ì—ì„œ ë¶ˆì¼ì¹˜ê°€ ë°œê²¬ë˜ì—ˆìŠµë‹ˆë‹¤. ìƒì„¸ ë‚´ìš©ì€ ìœ„ ë¡œê·¸ë¥¼ í™•ì¸í•˜ì„¸ìš”.")
```

### 3.1. ê²€ì¦ ì‹¤íŒ¨ ìƒì„¸ ëª©ë¡ (í•´ë‹¹í•˜ëŠ” ê²½ìš°)

ë§Œì•½ ê²€ì¦ì— ì‹¤íŒ¨í•œ íŒŒì¼ì´ ìˆë‹¤ë©´ ì•„ë˜ì— ëª©ë¡ì„ ì¶œë ¥í•©ë‹ˆë‹¤.

```python
df_results = pd.DataFrame(verification_results)
failed_scans = df_results[df_results['verified'] == False]

if not failed_scans.empty:
    print("--- ê²€ì¦ ì‹¤íŒ¨ ëª©ë¡ ---")
    print(failed_scans.to_string(index=False))
else:
    print("ëª¨ë“  í˜ì–´ ê²€ì¦ì— ì„±ê³µí•˜ì—¬ ì‹¤íŒ¨ ëª©ë¡ì´ ì—†ìŠµë‹ˆë‹¤.")
```
