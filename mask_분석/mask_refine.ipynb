# FastMRI – **Mask Archetype Discovery**
brain / knee × acc4 / acc8 의 k‑space 마스크를 정밀 분석하여
**‘정해진’ mask 함수와 파라미터**를 자동 추출‑제안해 주는 노트북입니다.

> **목표** 
> 1. 모든 파일에서 *줄 간격(Δidx)* 과 *중앙 low‑freq 폭* 을 계량화
> 2. 분포가 *단일 값*이면 그대로 고정, 다르면 `[min,max]` 범위 제안
> 3. mask 유형 판단 → **equispaced** 또는 **equispaced_fraction** 선택
> 4. 최종 `accel_cfgs` YAML 블록 출력 (DuplicateMaskDataset 용)

**사용법** 
1. 아래 `DATA_DIR` 변수만 본인 경로로 수정 → 전체 실행
2. Section 5 에 출력되는 YAML 을 `maskDuplicate/*.yaml` 로 복사
3. Section 6 시각화로 원본·복제 mask 가 겹치는지 확인


```python

import os, re, glob, json, warnings, math
import numpy as np, h5py
import matplotlib.pyplot as plt
from collections import defaultdict, Counter
from tqdm.auto import tqdm
warnings.filterwarnings("ignore")

# 수정: FastMRI 루트
DATA_DIR = "/home/swpants05/Desktop/2025_FastMri/Data"
```

```python

REG = re.compile(r"(brain|knee)_acc([48])_(\d+)\.h5")
LB_REG= re.compile(r"(brain|knee)_test(\d+)\.h5")

def list_files(split,dtype):
    if split in ("train","val"):
        return sorted(glob.glob(os.path.join(DATA_DIR,split,dtype,"*.h5")))
    if split=="leaderboard":
        return sorted(glob.glob(os.path.join(DATA_DIR,split,"acc*",dtype,"*.h5")))
    return []
files={s:{t:list_files(s,t) for t in ("image","kspace")} for s in ("train","val","leaderboard")}
print({k:{t:len(v) for t,v in d.items()} for k,d in files.items()})
```

```python

def load_mask(p):
    with h5py.File(p,'r') as f: return f['mask'][()].astype(bool)

def center_block_width(m: np.ndarray) -> int:
    """
    Args
        m : 1-D bool mask  (len = #readout columns)

    Returns
        중앙의 fully-sampled 구간 길이 (low-frequency block width)
    """
    N   = len(m)
    mid = N // 2                       # k-space 중심 열 (짝수 길이여도 OK)

    # --- 왼쪽 경계 -------------------------------
    left = mid
    while left - 1 >= 0 and m[left - 1]:
        left -= 1

    # --- 오른쪽 경계 -----------------------------
    right = mid
    # 짝수 길이면 mid 가 오른쪽 절반을 가리키므로 먼저 포함
    while right < N and m[right]:
        right += 1

    return right - left                # 폭 (개수)


def outer_spacing(m):
    nz=np.where(m)[0]
    # 간격 리스트
    spacings=np.diff(nz)
    # 첫번째 spacing이 주기가 될 가능성 높음
    return int(round(np.median(spacings)))

records=[]
for split in ("train","val","leaderboard"):
    for kp in tqdm(files[split]["kspace"], desc=split):
        m=load_mask(kp)
        fname=os.path.basename(kp)
        mm=REG.search(fname)
        if mm:
            body, acc, idx=mm.groups()
        else:
            mm=LB_REG.search(fname)
            if not mm: continue
            body, idx = mm.groups(); acc = "4" if "/acc4/" in kp else "8"
        records.append(dict(split=split, body=body, acc=acc,
                            N=len(m), centerW=center_block_width(m),
                            spacing=outer_spacing(m)))
print("parsed",len(records),"masks")
```

## 1️⃣  통계 요약

```python

import pandas as pd, itertools
df=pd.DataFrame(records)
summary = df.groupby(["body","acc"])[["centerW","spacing"]].agg(['mean','std','min','max'])
display(summary)
```

## 2️⃣  마스크 archetype 판정

```python

arche=[]   # (body,acc,mask_type,center_range,spacing)
for (b,a),grp in df.groupby(["body","acc"]):
    sp=set(grp['spacing'])
    cw_min,cw_max=int(grp.centerW.min()),int(grp.centerW.max())
    # brain 은 spacing 하나뿐 & cw 하나면 equispaced
    if len(sp)==1 and cw_min==cw_max:
        mtype="equispaced"
        center=cw_min
    else:
        mtype="equispaced_fraction"
        center=[cw_min, cw_max]
    arche.append(dict(body=b,acc=int(a),mask_type=mtype,
                      spacing=int(sorted(sp)[0]),
                      center=center))
arche
```

## 3️⃣  `accel_cfgs` YAML 제안

```python

yaml_lines=["accel_cfgs:"]
for d in arche:
    line=f"  - {{accel: {d['acc']}, mask_type: {d['mask_type']}, "
    if isinstance(d['center'],list):
        cf_low = round(d['center'][0]/d['spacing']/d['acc'],3)
        cf_hi  = round(d['center'][1]/d['spacing']/d['acc'],3)
        line+=f"cf: [{cf_low}, {cf_hi}]}}"
    else:
        cf = round(d['center']/d['spacing']/d['acc'],3)
        line+=f"cf: {cf}}}"
    yaml_lines.append(line)
print("\n".join(yaml_lines))
```

## 4️⃣  샘플 mask → 원본과 비교

```python

# pick first brain acc4 file & visualise
target=df[(df.body=='brain')&(df.acc=='4')].iloc[0]
m_orig=load_mask(files[target.split]['kspace'][0])

# --- duplicate sample ---
import torch
from duplicate_dataset import DuplicateMaskDataset
class Dummy(torch.utils.data.Dataset):
    def __len__(self): return 1
    def __getitem__(self,idx):
        return m_orig, np.zeros((8,256,256)), -1,-1,'dummy',0,'brain_x4'
dup=DuplicateMaskDataset(Dummy(), accel_cfgs=[dict(accel=4,cf=[0.1,0.12],mask_type='equispaced')])
m_dup=dup[0][0]

plt.figure(figsize=(6,2))
plt.subplot(1,2,1); plt.title('orig'); plt.plot(m_orig.astype(int))
plt.subplot(1,2,2); plt.title('dup'); plt.plot(m_dup.squeeze())
plt.tight_layout(); plt.show()
```

```python
# === KNEE 마스크 중심폭 분석 ===
import pandas as pd, seaborn as sns, matplotlib.pyplot as plt
sns.set_style("whitegrid")

knee_df = df[df.body == "knee"].copy()      # df 는 앞 셀에서 만든 데이터프레임
knee_df["center_frac"] = knee_df.centerW / knee_df.N

# 1) 요약 통계
display(
    knee_df
      .groupby(["acc"])[["N","centerW","center_frac"]]
      .agg(["mean","std","min","max","nunique"])
)

# 2) scatter  centerW vs N  (acc4/8 색 구분)
plt.figure(figsize=(4,3))
sns.scatterplot(data=knee_df, x="N", y="centerW", hue="acc", palette="tab10")
plt.title("knee – centerW vs N"); plt.show()

# 3) 히스토그램  center_fraction
g=sns.displot(data=knee_df, x="center_frac", hue="acc",
              kind="hist", bins=20, aspect=1.4)
g.fig.suptitle("knee – center_fraction histogram", y=1.03)
plt.show()

# 4) 테이블 :  (N, centerW) 빈도
pivot = knee_df.pivot_table(index="N", columns="centerW",
                            values="acc", aggfunc="size", fill_value=0)
print("빈도표  (행: readout N, 열: centerW)")
display(pivot)
```
