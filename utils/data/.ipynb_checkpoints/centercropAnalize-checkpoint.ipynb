# CenterCrop 검증 실험
- SSIM 차이(ΔSSIM) 측정: crop 적용 전후 비교
- Fully-sampled / Undersampled 모두 테스트
- 세 가지 crop 모드 비교: ZeroPad / NoisePad / K-space

```python
# 1) Setup & Imports
import os, re, glob, warnings, random
from pathlib import Path
from collections import defaultdict
import numpy as np
import matplotlib.pyplot as plt
import h5py
from skimage.metrics import structural_similarity as ssim
from tqdm.auto import tqdm
import torch
import torch.nn.functional as F
from typing import Tuple
from collections.abc import Mapping

warnings.filterwarnings("ignore", category=RuntimeWarning)
plt.rcParams['figure.dpi'] = 120
random.seed(42)

# 사용자 정의 모듈
from transforms import (
    CenterCropZeroPad,
    CenterCropNoisePad,
    KspaceCenterCropPad,
)
```

### 2. 경로 설정 & 파일 목록 생성

```python
# 2) 데이터 경로 & 파일 목록
DATA_DIR = Path('/home/swpants05/Desktop/2025_FastMri/Data')
REG     = re.compile(r"(brain|knee)_acc([48])_(\d+)\.h5")
LB_REG  = re.compile(r"(brain|knee)_test(\d+)\.h5")

def list_files(split: str, data_type: str):
    if split in ('train','val'):
        return sorted(glob.glob(str(DATA_DIR/split/data_type/'*.h5')))
    elif split=='leaderboard':
        out=[]
        for acc in ('acc4','acc8'):
            out+=sorted(glob.glob(str(DATA_DIR/'leaderboard'/acc/data_type/'*.h5')))
        return out
    else:
        raise ValueError
splits = ['train','val','leaderboard']
files = {sp:{'image':list_files(sp,'image'),
             'kspace':list_files(sp,'kspace')} for sp in splits}
```

### 3. Helper Functions

```python
# 3) 공통 헬퍼 함수
def quick_ifft2(kc):
    return np.fft.ifftshift(
        np.fft.ifft2(np.fft.fftshift(kc,axes=(-2,-1)),norm='ortho'),
        axes=(-2,-1)
    )

def rss(img, axis=0):
    return np.sqrt(np.sum(np.abs(img)**2,axis=axis))

def compute_recon(kspace, mask=None):
    if mask is not None:
        kspace = kspace * mask[np.newaxis,...]
    img_c = quick_ifft2(kspace)
    return rss(img_c,axis=0)

def compute_ssim(pred, ref):
    return ssim(pred, ref, data_range=ref.max()-ref.min())

def center_crop_or_pad(img: np.ndarray, target_shape: Tuple[int,int]) -> np.ndarray:
    h,w = img.shape; th,tw = target_shape
    # crop
    if h>th: img=img[(h-th)//2:(h-th)//2+th,:]
    if w>tw: img=img[:,(w-tw)//2:(w-tw)//2+tw]
    # pad
    h2,w2=img.shape; ph,pw=max(0,th-h2),max(0,tw-w2)
    img=np.pad(img,
               ((ph//2,ph-ph//2),(pw//2,pw-pw//2)),
               mode='constant',constant_values=0)
    return img
```

### 1) Baseline measurement (no crop vs GT / under vs GT)

```python
baseline = defaultdict(list)
for sp in splits:
    for img_p, ks_p in tqdm(zip(files[sp]['image'], files[sp]['kspace']),
                             total=len(files[sp]['image']),
                             desc=f"{sp} baseline"):
        with h5py.File(img_p,'r') as f:
            key='image_label' if 'image_label' in f else 'image_grappa'
            img_vol=f[key][()]
        with h5py.File(ks_p,'r') as f:
            ks_vol=f['kspace'][()]; mask=f['mask'][()]
        for gt, ks in zip(img_vol, ks_vol):
            rec_full  = compute_recon(ks)
            rec_under = compute_recon(ks, mask)
            # shape align
            rec_full  = center_crop_or_pad(rec_full, gt.shape)
            rec_under = center_crop_or_pad(rec_under, gt.shape)
            baseline[f"{sp}_full"].append (compute_ssim(rec_full, gt))
            baseline[f"{sp}_under"].append(compute_ssim(rec_under,gt))

# boxplot
plt.figure(figsize=(10,4))
keys=['train_full','train_under','val_full','val_under','leaderboard_full','leaderboard_under']
plt.boxplot([baseline[k] for k in keys],labels=keys,showmeans=True)
plt.xticks(rotation=45); plt.title("Baseline SSIM"); plt.tight_layout()
plt.show()
```

### 2)  Variants: ZeroPad / NoisePad / KspacePad

```python
# 5) Variants: ZeroPad / NoisePad / KspacePad
crop_size = (384,384)
variants = {
    'zero':  CenterCropZeroPad(crop_size),
    'noise': CenterCropNoisePad(crop_size,corner_size=8),
    'kspace':KspaceCenterCropPad(crop_size),
}

delta_records = defaultdict(list)
for sp in splits:
    for img_p, ks_p in tqdm(zip(files[sp]['image'], files[sp]['kspace']),
                             total=len(files[sp]['image']),
                             desc=f"{sp} variants"):
        with h5py.File(img_p,'r') as f_img:
            key='image_label' if 'image_label' in f_img else 'image_grappa'
            img_vol=f_img[key][()]
        with h5py.File(ks_p,'r') as f_ks:
            ks_vol=f_ks['kspace'][()]; mask=f_ks['mask'][()]
        for gt, ks in zip(img_vol, ks_vol):
            # baseline under
            rec_base = compute_recon(ks, mask)
            rec_base = center_crop_or_pad(rec_base, gt.shape)
            s_base   = compute_ssim(rec_base, gt)

            # test each variant
            mask_t = torch.from_numpy(mask.flatten()).float()
            ks_t   = torch.from_numpy(ks).to(torch.complex64)
            # for name, transformer in variants.items():
            #     m2, k2, *_ = transformer(mask_t, ks_t, None, {}, '', 0)
            #     arr = k2.numpy(); ks2 = arr[...,0]+1j*arr[...,1]
            #     rec = compute_recon(ks2)
            #     rec = center_crop_or_pad(rec, gt.shape)
            #     s_test = compute_ssim(rec, gt)
            #     delta_records[(sp,name)].append(s_test - s_base)
            
            for name, transformer in variants.items():
                # 1) transform → m2 (W0,), k2 (C, H0, W0, 2)
                m2, k2, *_ = transformer(mask_t, ks_t, None, {}, '', 0)
            
                # 2) Torch 상에서 mask 재적용
                #    m2: (W0,) → (1, 1, W0, 1) 으로 reshape
                m2_t = m2.view(1, 1, -1, 1)
                k2_masked = k2 * m2_t       # (C, H0, W0, 2), real/imag 유지
            
                # 3) NumPy 복합 k-space 생성
                arr = k2_masked.numpy()     # (C, H0, W0, 2)
                ks2 = arr[..., 0] + 1j * arr[..., 1]  # (C, H0, W0)
            
                # 4) Reconstruction & SSIM
                rec = compute_recon(ks2)
                rec = center_crop_or_pad(rec, gt.shape)
                s_test = compute_ssim(rec, gt)
            
                delta_records[(sp, name)].append(s_test - s_base)


# boxplot ΔSSIM
plt.figure(figsize=(8,4))
data=[delta_records[('train',v)] for v in variants]
plt.boxplot(data,labels=list(variants.keys()),showmeans=True)
plt.title("ΔSSIM (train under)"); plt.tight_layout(); plt.show()
```

```python
# boxplot ΔSSIM: split × variant
for sp in splits:
    plt.figure(figsize=(8,4))
    # 각 variant에 대응하는 split 데이터 꺼내기
    data = [delta_records[(sp, v)] for v in variants.keys()]
    plt.boxplot(data, labels=list(variants.keys()), showmeans=True)
    plt.title(f"ΔSSIM ({sp} undersampled) per variant")
    plt.ylabel("ΔSSIM")
    plt.xticks(rotation=0)
    plt.tight_layout()
    plt.show()
```

```python
# 5) Variants: ZeroPad / NoisePad / KspacePad
crop_size = (768,480)
variants = {
    'zero':  CenterCropZeroPad(crop_size),
    'noise': CenterCropNoisePad(crop_size,corner_size=8),
    'kspace':KspaceCenterCropPad(crop_size),
}

delta_records = defaultdict(list)
for sp in splits:
    # for img_p, ks_p in tqdm(zip(files[sp]['image'], files[sp]['kspace']),
    #                          total=len(files[sp]['image']),
    #                          desc=f"{sp} variants"):
    
    # 1) 파일 단위 20% 샘플링
    all_pairs = list(zip(files[sp]['image'], files[sp]['kspace']))
    sample_n  = max(1, int(len(all_pairs) * 0.2))
    sampled_pairs = random.sample(all_pairs, sample_n)
    for img_p, ks_p in tqdm(sampled_pairs,
                             total=sample_n,
                             desc=f"{sp} variants (files {sample_n}/{len(all_pairs)})"):

        with h5py.File(img_p,'r') as f_img:
            key='image_label' if 'image_label' in f_img else 'image_grappa'
            img_vol=f_img[key][()]
        with h5py.File(ks_p,'r') as f_ks:
            ks_vol=f_ks['kspace'][()]; mask=f_ks['mask'][()]
        # for gt, ks in zip(img_vol, ks_vol):
        # 2) 이 볼륨 내 슬라이스 단위로 20% 샘플링
        num_slices = img_vol.shape[0]
        slice_n    = max(1, int(num_slices * 0.2))
        slice_idxs = random.sample(range(num_slices), slice_n)
        for idx in slice_idxs:
            gt = img_vol[idx]
            ks = ks_vol[idx]
            # baseline under
            rec_base = compute_recon(ks, mask)
            rec_base = center_crop_or_pad(rec_base, gt.shape)
            s_base   = compute_ssim(rec_base, gt)

            # test each variant
            # fastMRI mask은 (H_k, W_k) 혹은 (W_k,) 형태인데,
            # 우리가 필요로 하는 것은 “column-wise” 1D 마스크 (W_k,)
            if mask.ndim == 2:
                # H_k x W_k 인 경우, 아무 row나 혹은 평균을 써도 column pattern은 동일
                mask_1d = mask[0]
            else:
                mask_1d = mask
            mask_t = torch.from_numpy(mask_1d).float()  # shape (W_k,)
            ks_t   = torch.from_numpy(ks).to(torch.complex64)
            # for name, transformer in variants.items():
            #     m2, k2, *_ = transformer(mask_t, ks_t, None, {}, '', 0)
            #     arr = k2.numpy(); ks2 = arr[...,0]+1j*arr[...,1]
            #     rec = compute_recon(ks2)
            #     rec = center_crop_or_pad(rec, gt.shape)
            #     s_test = compute_ssim(rec, gt)
            #     delta_records[(sp,name)].append(s_test - s_base)
            for name, transformer in variants.items():
                m2, k2, *_ = transformer(mask_t, ks_t, None, {}, '', 0)
            
                # 1) k2 차원에 따라 분기
                if k2.ndim == 4:
                    # real/imag last‐dim 있는 경우: (C, H0, W0, 2)
                    C, d1, d2, two = k2.shape
                    # W0은 crop_size[1]
                    # axis 1 또는 2 중 d? == W0 로 찾기
                    axis = 1 if d1 == crop_size[1] else 2
                    # view shape
                    shape = [1, 1, 1, 1]
                    shape[axis] = crop_size[1]
                    m2_t = m2.view(*shape)           # (1, W0, 1, 1) or (1,1,W0,1)
                    k2_masked = k2 * m2_t
            
                    # to numpy complex
                    arr = k2_masked.numpy()
                    ks2 = arr[..., 0] + 1j * arr[..., 1]
            
                elif k2.ndim == 3:
                    # real/imag 없고 복소 직접 담겼거나 DC‐injected 경우: (C, H0, W0)
                    C, d1, d2 = k2.shape
                    # width axis = 2
                    shape = [1, 1, crop_size[1]]
                    m2_t = m2.view(*shape)           # (1,1,W0)
                    k2_masked = k2 * m2_t
            
                    ks2 = k2_masked.numpy()          # 이미 complex dtype numpy
            
                else:
                    raise RuntimeError(f"Unexpected k2.ndim={k2.ndim}")
            
                # 2) Reconstruction & SSIM
                rec    = compute_recon(ks2)
                rec_sm = center_crop_or_pad(rec, gt.shape)
                s_test = compute_ssim(rec_sm, gt)
                delta_records[(sp, name)].append(s_test - s_base)
```

### 3) Crop-Size Spectrum : pad-only vs crop-happens 그룹 ΔSSIM

```python
# 6) Crop-size spectrum: pad-only vs crop-happens 그룹 ΔSSIM
import pandas as pd

records=[]
for sp in splits:
    for img_p, ks_p in zip(files[sp]['image'],files[sp]['kspace']):
        with h5py.File(img_p,'r') as f: key='image_label' if 'image_label' in f else 'image_grappa'; img_vol=f[key][()]
        with h5py.File(ks_p,'r') as f: ks_vol=f['kspace'][()]; mask=f['mask'][()]
        for gt, ks in zip(img_vol, ks_vol):
            Hk,Wk = ks.shape[1:]
            rec_base = compute_recon(ks, mask); rec_base=center_crop_or_pad(rec_base,gt.shape)
            s_base = compute_ssim(rec_base,gt)
            mask_t = torch.from_numpy(mask.flatten()).float()
            ks_t = torch.from_numpy(ks).to(torch.complex64)
            for name, transformer in variants.items():
                _, k2, *_ = transformer(mask_t, ks_t, None, {}, '', 0)
                arr=k2.numpy(); ks2=arr[...,0]+1j*arr[...,1]
                rec=compute_recon(ks2); rec=center_crop_or_pad(rec,gt.shape)
                s_test=compute_ssim(rec,gt)
                group='pad_only' if (Hk<=crop_size[0] and Wk<=crop_size[1]) else 'crop'
                records.append({'split':sp,'variant':name,'group':group,'delta':s_test-s_base})

df = pd.DataFrame(records)
summary = df.groupby(['variant','group'])['delta'].agg(['mean','std','count']).reset_index()
print(summary)

# bar plot
pivot = summary.pivot('variant','group','mean')
pivot.plot(kind='bar', figsize=(6,4))
plt.title("Mean ΔSSIM by variant & group"); plt.ylabel("ΔSSIM"); plt.tight_layout(); plt.show()
```
